```python
# Importando biblioteca pandas no Python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iptu_df = pd.read_excel("iptu-bairro.xlsx")
display(iptu_df.head())
```

```python
# Criar uma coluna "MEDIA" que tenha nela o valor médio de iptu naquele bairro
iptu_df['MEDIA'] = iptu_df['VALOR_TOTAL_LANCADO'] / iptu_df['QDE_LANCAMENTOS']
display(iptu_df.head())
```

```python
# Classificar os bairros como IPTU caro ou barato a depender se o valor esta acima da media da cidade
display(iptu_df.describe())

iptu_df['Status'] = np.where(iptu_df['MEDIA'] > 3888.396032, 'Caro', 'Barato')
display(iptu_df.head())
```

```python
# identificar o bairro com o iptu mais caro de Belo Horizonte
display(iptu_df['MEDIA'].max())
max_index = iptu_df['MEDIA'].idxmax()
display(iptu_df.loc[max_index])
```

```python
# identificar o bairro com o iptu mais barato de Belo Horizonte
display(iptu_df['MEDIA'].min())
min_index = iptu_df['MEDIA'].idxmin()
display(iptu_df.loc[min_index])
```

```python
#Salvando o dataframe em um novo arquivo personalizado
excel = "estudo_iptu.xlsx"
iptu_df.to_excel(excel,"IPTU")
```

```python
#Vamor personalizar a planilha Excel...
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font, Color, Alignment

#Vai abrir o Excel no openpyxl
wb = load_workbook(excel) # wb = Workbook

#Pegar a planilha certa... usando o Sheet Name (nome da planilha)
ws = wb['IPTU'] # Work Sheet -> planilha atual, ativa, de trabalho

#Vamos pintar o cabeçalho da tabela de "cinzinha"
cinzinha = PatternFill("solid", fgColor="CCCCCC")
coords = ['A1', 'B1', 'C1', 'D1', 'E1', 'F1']
for coord in coords:
  ws[coord].fill = cinzinha

#Onde o IPTU for caro pinta a fonte de vermelho e deixa negrito...
MAX_ROW = ws.max_row
num_linha = 2
while (num_linha <= MAX_ROW):
  coord = 'F'+str(num_linha)
  if ws[coord].value == 'Caro':
    ws[coord].font = Font(bold=True, color="FF0000")
  num_linha = num_linha + 1
#Salvar o Excel
wb.save(excel)
```
